```python
import SimpleITK as sitk
import os
import numpy as np

def get_image_sizes(paths):
    """
    Collect image sizes for all .nii.gz files in the given paths using SimpleITK.
    Args:
        paths (list): List of directory paths to search for .nii.gz files.
    Returns:
        sizes (list): List of image sizes (as tuples) from the .nii.gz files.
    """
    sizes = []
    for path in paths:
        for file in os.listdir(path):
            file_path = os.path.join(path, file)
            
            try:
                img = sitk.ReadImage(file_path)
                img_size = img.GetSize()
                print(file_path)
                print(img_size)
                sizes.append(img_size)  # GetSize returns a tuple
                
            except Exception as e:
                print(f"Error loading {file_path}: {e}")
                
    return sizes

def calculate_percentile(sizes, percentile=95):
    """
    Calculate the given percentile for a list of image sizes.
    Args:
        sizes (list): List of image sizes (tuples).
        percentile (float): Percentile to calculate.
    Returns:
        percentile_size (tuple): Percentile size as a tuple of dimensions.
    """
    sizes_array = np.array(sizes)
    # Calculate the percentile for each dimension
    percentile_size = np.percentile(sizes_array, percentile, axis=0).astype(int)
    return tuple(percentile_size)

# Paths to the directories containing .nii.gz files
paths = [r"/path/to/workspace/classification_model_originalimage/totalsegment_work/resampled/female_train_val_patches", r"/path/to/workspace/classification_model_originalimage/totalsegment_work/resampled/male_train_val_patches", r"/path/to/workspace/classification_model_originalimage/totalsegment_work/resampled/twenty_train_val_patches"]

# Get all image sizes
image_sizes = get_image_sizes(paths)


if image_sizes:
    # Calculate the 95th percentile size
    size_95_percentile = calculate_percentile(image_sizes, 95)
    print(f"The 95% size of the images is: {size_95_percentile}")
else:
    print("No .nii.gz files found in the specified paths.")
```
