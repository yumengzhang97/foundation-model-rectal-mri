```python
#"rename all images name to subject name in train_val_set folder"
import os
import pandas as pd

path = r'/path/to/workspace/classificationmodel_original_sagittal/total_work/normalized4/train2'
csv_path = r'/path/to/workspace/classificationmodel_original_sagittal/total_work/normalized4/train_labels.csv'
```

```python
df = pd.read_csv(csv_path)
```

```python
import SimpleITK as sitk

# images_folder = "/path/to/nii/images"
sizes_list = []

for filename in os.listdir(path):
    if filename.endswith(".nii.gz"):
        image_path = os.path.join(path, filename)
        image = sitk.ReadImage(image_path)
        
        size = image.GetSize()
        sizes_list.append(size)
        # print(f"Image: {filename}, Size: {size}")

unique_sizes = set(sizes_list)
print(f"\nTotal number of unique sizes: {len(unique_sizes)}")
print("Unique sizes:", unique_sizes)
```

```python
#rename
for filename in os.listdir(path):
    if filename.endswith(".nii.gz"):
        file_prefix = filename[:4]
        match = df[df['Subjects'].str.startswith(file_prefix)]

        if not match.empty:
            subject_name = match['Subjects'].values[0]

            new_filename = f"{subject_name}.nii.gz"
            old_file_path = os.path.join(path, filename)
            new_file_path = os.path.join(path, new_filename)

            os.rename(old_file_path, new_file_path)
            print(f"Renamed {filename} to {new_filename}")
        else:
            print(f"No match found for {filename}")

print("All matching files have been renamed.")
```
