```python
import SimpleITK as sitk
import numpy as np
import os

path = r'/path/to/workspace/classificationmodel_harmonizednew/sagittal_harmonize/rectum_centrecrop/train'
new_path = r'/path/to/workspace/classificationmodel_harmonizednew/sagittal_harmonize/rectum_centrecrop/train2'
```

```python
for i in os.listdir(path):
    path2 = os.path.join(path, i)
    image = sitk.ReadImage(path2) #x,y,z= 16,224,224
    
    image_array = sitk.GetArrayFromImage(image) #z,y,x=192,192,22
    # print("old image_array", image_array.shape)
    
    image_array2 = np.transpose(image_array, (2,1,0)) #change from 48,192,192 -> 192,192,48
    # print("new image_array", image_array2.shape)

    image_new = sitk.GetImageFromArray(image_array2)
    image_new.CopyInformation(image_new)
    
    new_path2 = os.path.join(new_path, i)
    sitk.WriteImage(image_new, new_path2)

print('done')
```
