```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, recall_score

# Read CSV file
df = pd.read_csv(r"/path/to/workspace/Complete_set/Experiments/Results/FMmodel192/test/best_FM_192.csv")

# Extract ground truth and predicted probabilities for EVI and MFI
y_true_evi = df['ground_truth_evi'].values
y_pred_evi = df['predicted_probability_evi'].values

y_true_mfi = df['ground_truth_mfi'].values
y_pred_mfi = df['predicted_probability_mfi'].values

# Convert predicted probabilities to binary predictions
y_pred_binary_evi = (y_pred_evi > 0.5).astype(int)
y_pred_binary_mfi = (y_pred_mfi > 0.5).astype(int)

# Function to calculate AUC, Sensitivity, and Specificity
def calculate_metrics(y_true, y_pred, y_pred_binary):
    auc = roc_auc_score(y_true, y_pred)
    sensitivity = recall_score(y_true, y_pred_binary)  # Sensitivity (Recall for class 1)
    specificity = recall_score(y_true, y_pred_binary, pos_label=0)  # Specificity (Recall for class 0)
    return auc, sensitivity, specificity

# Function to compute 95% confidence intervals using Bootstrap method
def bootstrap_ci(y_true, y_pred, n_bootstraps=1000, seed=42):
    rng = np.random.RandomState(seed)
    aucs, sensitivities, specificities = [], [], []

    for _ in range(n_bootstraps):
        # Random sampling with replacement
        indices = rng.randint(0, len(y_true), len(y_true))
        if len(np.unique(y_true[indices])) < 2:  # Ensure the sample contains both classes (0 and 1)
            continue

        # Compute metrics for the bootstrap sample
        y_true_boot = y_true[indices]
        y_pred_boot = y_pred[indices]
        y_pred_binary_boot = (y_pred_boot > 0.5).astype(int)

        auc, sens, spec = calculate_metrics(y_true_boot, y_pred_boot, y_pred_binary_boot)
        aucs.append(auc)
        sensitivities.append(sens)
        specificities.append(spec)

    # Function to compute the 95% confidence interval
    def compute_ci(metric_list):
        lower = np.percentile(metric_list, 2.5)
        upper = np.percentile(metric_list, 97.5)
        return (lower, upper)

    return {
        'auc_ci': compute_ci(aucs),
        'sensitivity_ci': compute_ci(sensitivities),
        'specificity_ci': compute_ci(specificities)
    }

# Compute metrics and confidence intervals for both EVI and MFI
metrics_evi = bootstrap_ci(y_true_evi, y_pred_evi)
metrics_mfi = bootstrap_ci(y_true_mfi, y_pred_mfi)

# Print results
print("EVI 95% Confidence Intervals:", metrics_evi)
print("MFI 95% Confidence Intervals:", metrics_mfi)


```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, recall_score, f1_score, balanced_accuracy_score

# Read CSV file
df = pd.read_csv(r"/path/to/workspace/Complete_set/Experiments/Results/Segresnet/best_segresnet_axial_harmo.csv")

# Extract ground truth and predicted probabilities for EVI and MFI
y_true_evi = df['ground_truth_evi'].values
y_pred_evi = df['predicted_probability_evi'].values

y_true_mfi = df['ground_truth_mfi'].values
y_pred_mfi = df['predicted_probability_mfi'].values

# Convert predicted probabilities to binary predictions
y_pred_binary_evi = (y_pred_evi > 0.5).astype(int)
y_pred_binary_mfi = (y_pred_mfi > 0.5).astype(int)

# Function to calculate AUC, Sensitivity, Specificity, F1 Score, and Balanced Accuracy
def calculate_metrics(y_true, y_pred, y_pred_binary):
    auc = roc_auc_score(y_true, y_pred)
    sensitivity = recall_score(y_true, y_pred_binary)  # Sensitivity (Recall for class 1)
    specificity = recall_score(y_true, y_pred_binary, pos_label=0)  # Specificity (Recall for class 0)
    f1 = f1_score(y_true, y_pred_binary)  # F1 Score
    balanced_acc = balanced_accuracy_score(y_true, y_pred_binary)  # Balanced Accuracy
    return auc, sensitivity, specificity, f1, balanced_acc

# Function to compute 95% confidence intervals using Bootstrap method
def bootstrap_ci(y_true, y_pred, n_bootstraps=1000, seed=42):
    rng = np.random.RandomState(seed)
    aucs, sensitivities, specificities, f1_scores, balanced_accs = [], [], [], [], []

    for _ in range(n_bootstraps):
        # Random sampling with replacement
        indices = rng.randint(0, len(y_true), len(y_true))
        if len(np.unique(y_true[indices])) < 2:  # Ensure the sample contains both classes (0 and 1)
            continue

        # Compute metrics for the bootstrap sample
        y_true_boot = y_true[indices]
        y_pred_boot = y_pred[indices]
        y_pred_binary_boot = (y_pred_boot > 0.5).astype(int)

        auc, sens, spec, f1, balanced_acc = calculate_metrics(y_true_boot, y_pred_boot, y_pred_binary_boot)
        aucs.append(auc)
        sensitivities.append(sens)
        specificities.append(spec)
        f1_scores.append(f1)
        balanced_accs.append(balanced_acc)

    # Function to compute the 95% confidence interval
    def compute_ci(metric_list):
        lower = np.percentile(metric_list, 2.5)
        upper = np.percentile(metric_list, 97.5)
        return (lower, upper)

    return {
        'auc_ci': compute_ci(aucs),
        'sensitivity_ci': compute_ci(sensitivities),
        'specificity_ci': compute_ci(specificities),
        'f1_score_ci': compute_ci(f1_scores),
        'balanced_acc_ci': compute_ci(balanced_accs)
    }

# Compute metrics and confidence intervals for both EVI and MFI
metrics_evi = bootstrap_ci(y_true_evi, y_pred_evi)
metrics_mfi = bootstrap_ci(y_true_mfi, y_pred_mfi)

# Print results
print("EVI 95% Confidence Intervals:", metrics_evi)
print("MFI 95% Confidence Intervals:", metrics_mfi)
```
