```python
path_train = r"/path/to/workspace/nnunet_dataset/all_image/imagesTr"
new_path = r'/path/to/workspace/nnunet_dataset/transverse/imagesTr'
```

```python
import glob
import os
import shutil
```

```python
nifti_images = glob.glob(os.path.join(path_train, "*.nii.gz"))
# print(nifti_images)
print(len(nifti_images))
```

```python
print(nifti_images[:3])
```

```python
# Create a dictionary to store files for each patient
patients_files = {}

# Iterate through all found .nii.gz files
for file_path in nifti_images:
    filename = os.path.basename(file_path)
    
    # Assume naming convention as subjectname_path_number_filename.nii.gz, split by "_"
    parts = filename.split('_')
    
    # Get subjectname and path to form a unique identifier for the patient
    subject_path_id = '_'.join(parts[0:2])  # The first two parts identify the patient
    
    # Create a list for patient files
    if subject_path_id not in patients_files:
        patients_files[subject_path_id] = []
    
    # Add the file path to the patient's file list
    patients_files[subject_path_id].append(file_path)
```

```python
# print(patients_files)
# print(type(patients_files))
```

```python
# Process each patient
for subject_path_id, file_list in patients_files.items():
    # Split files into transverse (horizontal) and sagittal (vertical)
    transverse_files = []
    sagital_files = []
    
    for file_path in file_list:
        filename = os.path.basename(file_path)
        
        # Identify sagittal files (those with 'sag' in the filename) and categorize as vertical
        if 'sag' in filename:
            sagital_files.append(file_path)
            
        elif 'harmonization' not in filename:
            # All other files (with 'ax' or without 'ax') are categorized as transverse (horizontal)
            transverse_files.append(file_path) 

    # After excluding sagittal files, the remaining ones are transverse
    for transverse_file in transverse_files:
        # Copy transverse files to the new path
        shutil.copy(transverse_file, new_path)
        print(f"Copied: {transverse_file} to {new_path}")
```
