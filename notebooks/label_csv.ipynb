# Export labels by split


```python
import csv
import json
import os
import pandas as pd

DATA_SPLIT = "train"  # options: "train", "val", "test"

SPLIT_CONFIG = {
    "train": {
        "dataset_dir": r"/path/to/datasets/DATASET_UUID_REDACTED",
        "save_dir": r"/path/to/workspace/classification_model/after_centrecrop",
        "csv_name": "train_labels.csv",
    },
    "val": {
        "dataset_dir": r"/path/to/datasets/DATASET_UUID_REDACTED",
        "save_dir": r"/path/to/workspace/classification_model/after_centrecrop",
        "csv_name": "val_labels.csv",
    },
    "test": {
        "dataset_dir": r"/path/to/datasets/DATASET_UUID_REDACTED",
        "save_dir": r"/path/to/workspace/classification_model/after_centrecrop",
        "csv_name": "test_labels.csv",
    },
}

config = SPLIT_CONFIG[DATA_SPLIT]
eforms_file_path = os.path.join(config["dataset_dir"], "eforms.json")
with open(eforms_file_path) as f:
    studies = json.load(f)
```

```python
rows = []
for item in studies:
    subject = item.get("subject") or item.get("Subjects") or item.get("id")
    label = item.get("label") or item.get("Label") or item.get("Sex") or item.get("Gender")
    rows.append({"Subjects": subject, "Label": label})

os.makedirs(config["save_dir"], exist_ok=True)
output_path = os.path.join(config["save_dir"], config["csv_name"])
pd.DataFrame(rows).to_csv(output_path, index=False)
print(output_path)
```
