```python
import pandas as pd
from sklearn.model_selection import train_test_split

labels_csv_path = r"/path/to/workspace/classification_model/after_centrecropnew/train_val_labels.csv"
labels_df = pd.read_csv(labels_csv_path)
train_df, val_df = train_test_split(labels_df, test_size=0.2, random_state=42)

train_csv_path = "/path/to/workspace/classification_model/after_centrecropnew/split_train_val/train_labels.csv"
val_csv_path = "/path/to/workspace/classification_model/after_centrecropnew/split_train_val/val_labels.csv"
train_df.to_csv(train_csv_path, index=False)
val_df.to_csv(val_csv_path, index=False)

print(f"Training and validation sets saved to {train_csv_path} and {val_csv_path}")
```
