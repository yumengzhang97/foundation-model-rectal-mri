```python
import os
import shutil
import pandas as pd

def copy_files_based_on_csv(csv_file, source_folder, destination_folder):
    # ??CSV??
    data = pd.read_csv(csv_file)

    # ?????????
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # ??CSV????Subjects?
    for subject in data['Subjects']:
        prefix = subject[:4]

        for f in os.listdir(source_folder):
             if f.startswith(prefix):
                 matched_file = f
                 source_file = os.path.join(source_folder, matched_file)
        
                 # new_filename = subject + ".nii.gz"
                 destination_file = os.path.join(destination_folder, matched_file)

                 shutil.copy2(source_file, destination_file)
                 print(f"copy file to {destination_file}")


# ????
csv_file = csv_path1 # ?????CSV????
source_folder = path # ???????????
destination_folder = new_path1 # ????????????

copy_files_based_on_csv(csv_file, source_folder, destination_folder)

print('done')
```
