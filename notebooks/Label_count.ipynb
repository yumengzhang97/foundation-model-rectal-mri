```python
import os
datasets_dir_path = "/path/to/datasets"
datasets = os.listdir(datasets_dir_path)
datasets
```

```python
dataset = datasets[0]
dataset_dir_path = os.path.join(datasets_dir_path,dataset)
print(dataset_dir_path)
```

```python
import json
eforms_file_path = os.path.join(dataset_dir_path, "eforms.json")
with open(eforms_file_path) as f:
    studies = json.load(f)
```

```python
studies[1]['eForm']['chaimeleon_patient_id']
```

```python
EVI_positive = []
EVI_negative = []
MFI_positive = []
MFI_negative = []
for i in range(0,len(studies)):
    patient_id = studies[i]['eForm']['chaimeleon_patient_id']    
    EVI_label = studies[i]["eForm"]['pages'][2]['page_data']['extramural_invation_mr']['value']
    MFI_label = studies[i]["eForm"]['pages'][2]['page_data']['mesorectal_invation_mr']['value']
    
    if EVI_label == True:
        EVI_positive.append(patient_id)
    elif EVI_label == False:
        EVI_negative.append(patient_id)

    if MFI_label == True:
        MFI_positive.append(patient_id)
    elif MFI_label == False:
        MFI_negative.append(patient_id)
        
print('number of EVI+ :', len(EVI_positive))
# print(EVI_positive)
print('number of EVI- :', len(EVI_negative))
# print(EVI_negative)
print('number of MFI+ :', len(MFI_positive))
print('number of MFI- :', len(MFI_negative))
```
