```python
import glob
import os
```

```python
path_test = r"/path/to/workspace/classification_model_originalimage/test_data"
path_data_test = r"/path/to/workspace/test_dataset_nifti"

nifti_images = glob.glob(os.path.join(path_data_test, "*", "*", "*", "*.nii.gz"))
print(nifti_images)
```

```python
# import shutil as sh
from tqdm import tqdm 

patients_files = {}
for img_tmp in tqdm(nifti_images):
    print(img_tmp.split("/"))
    subject_path_id = img_tmp.split("/")[5]
    print(subject_path_id)
    # Create a list for patient files
    if subject_path_id not in patients_files:
        patients_files[subject_path_id] = []
        
    patients_files[subject_path_id].append(img_tmp)
    
    # save_pth = os.path.join(path_train, img_tmp.split("/")[-4] + "_" + img_tmp.split("/")[-3] + "_" + img_tmp.split("/")[-2] + "_" +  img_tmp.split("/")[-1].split(".nii.gz")[0] + "_0000.nii.gz")
    # print(save_pth)
    # sh.copyfile(img_tmp, save_pth)
```

```python
import shutil as sh
# Process each patient
for subject_path_id, file_list in patients_files.items():
    # Split files into transverse (horizontal) and sagittal (vertical)
    transverse_files = []
    sagital_files = []
    
    for file_path in file_list:
        filename = os.path.basename(file_path)
        
        # Identify sagittal files (those with 'sag' in the filename) and categorize as vertical
        if 'sag' in filename:
            sagital_files.append(file_path)
            
        elif 'harmonization' not in filename:
            # All other files (with 'ax' or without 'ax') are categorized as transverse (horizontal)
            transverse_files.append(file_path) 

    # After excluding sagittal files, the remaining ones are transverse
    for transverse_file in transverse_files:
        # Copy transverse files to the new path
        # base_path = os.path.dirname(transverse_file)
        # harmonize_file = os.path.join(base_path, 'harmonization_sample.nii.gz')
        # print(transverse_file)
        save_pth = os.path.join(path_test, transverse_file.split("/")[5][:4] + "_" + transverse_file.split("/")[7] + "_" + transverse_file.split("/")[8])
        # print(save_pth)
        sh.copy(transverse_file, save_pth)
        print(f"Copied: {transverse_file} to {save_pth}")
        
print("done") 
```
