```python
import os
import shutil

# ???B???N2???
path_B = "/path/to/shared/GROUPS/UM/rectum/test_dataset_nifti"  # ???????
# ????C??????????
path_C = "/path/to/shared/ALL/UM_to_QUIBIM_Harmonization/rectum/axial"  # ???????
# ????
path_A = "/path/to/workspace/classification_model/test_data"  # N1??????



def copy_files(path_A, path_B, path_C):
    # ????A???????????A??????
    for filename in os.listdir(path_A):
        if filename.endswith(".nii.gz"):
            # ?????ID
            parts = filename.split("_")
            prefix = parts[0]
            id = parts[1]
            
            # ????????????B + ???????? + ??????? + ID???
            target_dir = os.path.join(path_B, f"{prefix}*", "*", id)
            
            # ??glob??????????????????
            import glob
            matched_dirs = glob.glob(target_dir)
            
            for dir_path in matched_dirs:
                # ???????????
                for file in os.listdir(dir_path):
                    if file.endswith(".nii.gz") and "harmonization" not in file:
                        src = os.path.join(dir_path, file)
                        dst = os.path.join(path_C, f"{prefix}_{id}.nii.gz")
                        shutil.copy(src, dst)
                        print(f"Copied: {src} -> {dst}")

# ????
copy_files(path_A, path_B, path_C)
print('done')
```
