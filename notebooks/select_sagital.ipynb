```python
import os
import nibabel as nib
import numpy as np

# ?????????
input_dir = r"/path/to/workspace/classification_model/val_data"  # ??????????
output_file = "sagittal_files_val.txt"    # ?? sagittal ??????
```

```python
sagittal_files = []

def is_sagittal(affine):
    """
    ????? sagittal??????
    """
    # ?? affine ??????????
    orientation = np.abs(affine[:3, :3])  # ?????????
    slice_direction = np.argmax(orientation[:, 2])  # ???????
    
    # sagittal ???????? X ???
    return slice_direction == 0
```

```python
for root, dirs, files in os.walk(input_dir):
    for file in files:
        if file.endswith(".nii") or file.endswith(".nii.gz"):
            nii_path = os.path.join(root, file)
            try:
                # ?? NIfTI ???
                nii_img = nib.load(nii_path)
                affine = nii_img.affine  # ?? affine ??
                
                # ????? sagittal
                if is_sagittal(affine):
                    sagittal_files.append(nii_path)
            except Exception as e:
                print(f"Error processing file {nii_path}: {e}")

# ? sagittal ?????????
with open(output_file, "w") as f:
    for filename in sagittal_files:
        f.write(filename + "\n")

print(f"Sagittal files saved to {output_file}")
```
