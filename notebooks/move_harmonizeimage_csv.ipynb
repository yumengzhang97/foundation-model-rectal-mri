```python
import os
import shutil
import pandas as pd

# --- 1) Define file paths and read CSVs ---
train_csv = "/path/to/workspace/classificationmodel_harmonizednew/train_labels.csv"
val_csv   = "/path/to/workspace/classificationmodel_harmonizednew/val_labels.csv"
test_csv  = "/path/to/workspace/classificationmodel_harmonizednew/test_labels.csv"

df_train = pd.read_csv(train_csv)
df_val   = pd.read_csv(val_csv)
df_test  = pd.read_csv(test_csv)

# Convert 'Subjects' to string (just in case) and create sets
train_subjects = set(df_train["Subjects"].astype(str).str[:4])
val_subjects   = set(df_val["Subjects"].astype(str).str[:4])
test_subjects  = set(df_test["Subjects"].astype(str).str[:4])

# --- 2) Define source folder and destination folders ---
source_folder = "/path/to/shared/ALL/UM_to_QUIBIM_Harmonization/rectum/sagittal"
train_folder  = "/path/to/workspace/classificationmodel_original_sagittal/train"
val_folder    = "/path/to/workspace/classificationmodel_original_sagittal/val"
test_folder   = "/path/to/workspace/classificationmodel_original_sagittal/test"

os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# --- 3) Move images based on subject ID (the 'e17d' part) ---
# Example filename: "harmonized_SUBJECT_ID_REDACTED.nii.gz"
# Splitting on '_' -> ["harmonized", "e17d", "8278.nii.gz"]
# We want the second to last element: parts[-2] = "e17d"

all_files = os.listdir(source_folder)

for filename in all_files:
    # Skip directories
    if os.path.isdir(os.path.join(source_folder, filename)):
        continue
    
    # Only handle .nii.gz files
    if not filename.endswith(".nii.gz"):
        print(f"Skipping file that doesn't end with .nii.gz: {filename}")
        continue

    # parts = filename.split("_")
    
    # if len(parts) < 3:
    #     print(f"Warning: filename {filename} doesn't match expected pattern.")
    #     continue
    
    # # The second to last element is "e17d"
    # subject_id = parts[-2]  # "e17d"

    subject_id = filename.split("_")[0]
    
    # Decide which folder to move to based on the subject sets
    if subject_id in train_subjects:
        dest_path = os.path.join(train_folder, filename)
        shutil.copy(os.path.join(source_folder, filename), dest_path)
    elif subject_id in val_subjects:
        dest_path = os.path.join(val_folder, filename)
        shutil.copy(os.path.join(source_folder, filename), dest_path)
    elif subject_id in test_subjects:
        dest_path = os.path.join(test_folder, filename)
        shutil.copy(os.path.join(source_folder, filename), dest_path)
    else:
        print(f"Warning: {filename} (subject {subject_id}) not found in any CSV list.")
        
print('done')
```
