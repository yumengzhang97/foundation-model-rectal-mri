```python
import csv
save_dir = r'/path/to/workspace/classification_model_originalimage/totalsegment_work/resampled'
csv_name = "Gender_find.csv"
csv_file_path = os.path.join(save_dir, csv_name)
```

```python
studies[0]['eForm']['chaimeleon_patient_id']
```

```python
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Subjects', 'Dataset', 'Gender'])

    for i in range(0,len(studies)):
        subject_name = studies[i]['eForm']['chaimeleon_patient_id']
        dataset = 'train'
        gender = studies[i]['eForm']['pages'][1]['page_data']['gender']['value']

        writer.writerow([subject_name, dataset, gender])

    print("train done")
    
    for j in range(0,len(studies2)):
        subject_name2 = studies2[j]['eForm']['chaimeleon_patient_id']
        dataset2 = 'val'
        gender2 = studies2[j]['eForm']['pages'][1]['page_data']['gender']['value']
        
        writer.writerow([subject_name2, dataset2, gender2])
    

print(f"Data saved to {csv_file_path}")
```
