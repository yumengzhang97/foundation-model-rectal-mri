```python
import SimpleITK as sitk
import glob
import os

def collect_image_orientations(image_dir, extension=".nii.gz"):
    """
    Collect and print the orientation (Direction Cosines) of MR images in a directory.

    :param image_dir: Path to the directory containing MR images.
    :param extension: File extension of the images (default is '.nii.gz').
    """
    # Get all image paths
    image_paths = glob.glob(os.path.join(image_dir, f"*{extension}"))
    
    # Dictionary to store image paths and their orientations
    orientation_dict = {}
    
    for img_path in image_paths:
        try:
            
            # Read the image
            image = sitk.ReadImage(img_path)
            
            # Get the Direction Cosines (orientation matrix)
            direction = image.GetDirection()  # Returns a flat tuple
            
            # Store orientation
            orientation_dict[img_path] = direction
            
            # Print the orientation
            print(f"Image: {os.path.basename(img_path)}")
            print(f"Orientation (Direction Cosines): {direction}\n")
        
        except Exception as e:
            print(f"Error reading {img_path}: {e}")

    return orientation_dict

if __name__ == "__main__":
    # Example usage
    image_directory = r"/path/to/workspace/classification_model_originalimage/train_data"
    orientations = collect_image_orientations(image_directory, extension=".nii.gz")
    
    # Optional: Find unique orientations
    unique_orientations = set(orientations.values())
    print("\nUnique Orientations Found:")
    for idx, orient in enumerate(unique_orientations):
        print(f"{idx + 1}: {orient}")
```

```python
import os
import nibabel as nib

def print_nifti_orientations(nifti_root):
    """
    Recursively walk `nifti_root` for NIfTI files (.nii / .nii.gz),
    load each file using nibabel, and print out its orientation codes.
    """
    for dirpath, dirnames, filenames in os.walk(nifti_root):
        for fname in filenames:
            if fname.lower().endswith('.nii') or fname.lower().endswith('.nii.gz'):
                fpath = os.path.join(dirpath, fname)
                try:
                    img = nib.load(fpath)
                    affine = img.affine

                    # Get orientation codes (e.g. ("R","A","S"))
                    axcodes = nib.aff2axcodes(affine)
                    
                    print(f"File: {fpath}")
                    print(f"  Orientation codes: {axcodes}\n")

                except Exception as e:
                    # If it's not a valid NIfTI or can't be read, just skip
                    print(f"Could not read file: {fpath}, error: {e}")

if __name__ == "__main__":
    # Update this path to the folder containing your .nii/.nii.gz data
    nifti_folder = r"/path/to/workspace/classification_model_originalimage/train_data"
    print_nifti_orientations(nifti_folder)
```
