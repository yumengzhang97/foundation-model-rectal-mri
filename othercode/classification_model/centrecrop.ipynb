```python
def center_crop(img, crop_size):
    size = img.GetSize()  # (width, height, depth)
    center = [int(size[i] / 2) for i in range(3)]   
    start = [max(0, min(center[i] - crop_size[i] // 2, size[i] - crop_size[i])) for i in range(3)]
    adjusted_crop_size = [min(crop_size[i], size[i] - start[i]) for i in range(3)]
    # start = [max(0, s) for s in start]
    # end = [min(start[i] + crop_size[i], original_size[i]) for i in range(3)] 
    # cropped_img = img[start[0]:end[0], start[1]:end[1], start[2]:end[2]]
    
    cropped_img = sitk.RegionOfInterest(img, adjusted_crop_size, start)
    
    return cropped_img
```

```python
import SimpleITK as sitk
import numpy as np

new_dir = r'/home/chaimeleon/persistent-home/classification_model_originalimage/totalsegment_work/resampled/centrecrop/train_val'
crop_size = (224, 224, 48)


for img_path in image_paths:
    img = sitk.ReadImage(img_path)
    
    file_name = img_path.split('/')[-2] + '.nii.gz' #patient number
    output_path = os.path.join(new_dir, file_name)
    
    cropped_img = center_crop(img, crop_size)

    sitk.WriteImage(cropped_img, output_path)
    print(f"Cropped image saved to {output_path}")
```
